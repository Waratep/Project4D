```python
import cv2
import numpy as np
from scipy import ndimage as ndi
from skimage.morphology import square,dilation
```

```python
def region_of_interest(img, vertices):   
    mask = np.zeros_like(img)
    match_mask_color = (255,)
    cv2.fillPoly(mask, vertices, 255)
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image

def slope(start,end,width):
    x1=start[0]
    x2=end[0]
    y1=width-start[1]
    y2=width-end[1]
#     print('(x2-x1):',(x2-x1))
    if (x2-x1) != 0:
        m=(y2-y1)/(x2-x1)        
        return m
```

```python
url='../video/test.h264'
cap = cv2.VideoCapture(url)
count=0
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        print("Can't receive frame (stream end?). Exiting ...")
        break
        
    count+=1    
    # resize image
    frame = cv2.resize(frame,(300, 200))    
    height=frame.shape[0]
    width=frame.shape[1]
    region_of_interest_vertices = [   
        (0, height),
        (0, (height/3)),
        (width, (height/3)),
        (width, height)]
    
    # rgb to gray
    gray=cv2.cvtColor(frame, cv2.COLOR_RGB2GRAY)

    # edge detection + dilation + gassian
    edges = cv2.Canny(gray,100,200)
    edges = dilation(edges, square(3))
    edges[:,width-3:width] = 0 
    edges = ndi.gaussian_filter(edges,2)
    
    # crop image
    crop = region_of_interest(edges,np.array([region_of_interest_vertices], np.int32))
    frame=cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)    
    lowerR=np.array([220,70,70])
    upperR=np.array([255,160,170])    
    maskR = cv2.inRange(frame, lowerR, upperR)

 
    countR=0
    for i in range(width):        
        if maskR[int(height/2)][i] == 255: 
            countR+=1
        if countR > 10:
            print()
            print(count,'stop')
            print()
            cv2.putText(frame,'stop',(190,25), font, 0.5,(0,0,0),2,cv2.LINE_AA)
            break

    # Hough Transform
    lines = cv2.HoughLinesP(crop,rho = 1,theta = 1*np.pi/180,threshold = 155,minLineLength = 10,maxLineGap = 250)
    font = cv2.FONT_HERSHEY_SIMPLEX
    l=[]
    r=[]
    
    # Draw lines on the image
    if(lines is not None) :
        for line in lines:  
            x1, y1, x2, y2 = line[0]
            if(slope((x1,y1),(x2,y2),width)):
                if(x1 < int(width/2)):
                    l.append([x1,x2,y1,y2,slope((x1,y1),(x2,y2),width)])
                else:
                    r.append([x1,x2,y1,y2,slope((x1,y1),(x2,y2),width)])
        
        l=np.array(l)
        r=np.array(r)
        
        # slope left
        if(len(l)):
            slope_left = round(sum(l[:,4])/len(l),2)  
            x1 = round(sum(l[:,0])/len(l)).astype('int')
            x2 = round(sum(l[:,1])/len(l)).astype('int')
            y1 = round(sum(l[:,2])/len(l)).astype('int')
            y2 = round(sum(l[:,3])/len(l)).astype('int')
            cv2.line(frame, (x1, y1), (x2, y2), (255, 0, 0), 3)
        else:
            slope_left = 0
        cv2.putText(frame,str(slope_left),(10,25), font, 0.5,(255,255,255),2,cv2.LINE_AA)
#         print('slope_left:', slope_left)

        # slope right
        if(len(r)):
            slope_right = round(sum(r[:,4])/len(r),2)
            x1 = round(sum(r[:,0])/len(r)).astype('int')
            x2 = round(sum(r[:,1])/len(r)).astype('int')
            y1 = round(sum(r[:,2])/len(r)).astype('int')
            y2 = round(sum(r[:,3])/len(r)).astype('int')
            cv2.line(frame, (x1, y1), (x2, y2), (0, 0, 255), 3)
        else:
            slope_right = 0
        cv2.putText(frame,str(slope_right),(100,25), font, 0.5,(255,255,255),2,cv2.LINE_AA)
        print('slope_right:', slope_right)  

#         different of slope
        print('slope: ', round(slope_right+slope_left,3))
        print()
        
    
    frame=cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)    
    cv2.imshow('frame', frame)
    if cv2.waitKey(1) == ord('q'):        
        break

cap.release()
cv2.destroyAllWindows()
    
```
